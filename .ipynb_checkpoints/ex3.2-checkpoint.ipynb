```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go

%matplotlib inline
```

```python
f = lambda t: np.exp(-t**2)

dt = 0.5    # sampling rate in seconds
t = np.arange(0,5.1,dt)    # sampling time points
xn = f(t)    # discrete signal, signal value for every point in time

plt.figure(figsize=(16,9))
t1 = np.linspace(0,5, 10000)    # t for continuous ideal plotting
plt.plot(t1, f(t1), label="Ideal function")
plt.scatter(t, xn, label="Samples", c="red")

plt.legend()
plt.show()
```

```python
def sinc_interp(t, signal, dt, m=None):
    
    """
    Perfroms a sinc interpolation
    
    Parameters
    ----------
    t: float, array-like
        Desired time at which we want to estimate the original signal
    
    signal: array-like
    
    dt: float
        Sampling rate
        
    m: int, default None
        Number of points for the calculation of sinc interpolation.
        If None, it will take all available points. Notice that m cannot be larger
        than the signal size
    """
    
    sinc = lambda t, i: np.sinc( (np.pi*(t - i*dt)) / dt )
    
    if m is None:
        m = signal.size
    
#     return [signal[i] * sinc(t, i) for i in range(m)]
    return sum([signal[i] * sinc(t, i) for i in range(m)])
```

```python
ti = 2.7    # the point we want to interpolate

f(ti)
```

```python
sinc_interp(ti, xn, dt, 5)
```

```python
sinc_interp(ti, xn, dt, 7)
```

```python
sinc_interp(ti, xn, dt, 9)
```
