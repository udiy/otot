```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go

%matplotlib inline
```

```python
def sinc_interp(t, xn, dt):
    """
    Perfroms a sinc interpolation
    
    Parameters
    ----------
    xn: array-like
        A list of numbers representing the signal value at certain points
    
    t: float, array-like
        Desired time at which we want to estimate the original signal
    
    dt: float
        Sampling rate
    """
    n = len(xn)
    n_vec = np.arange(n)
    
    # tile t vector to make 2 dimensional
    t = np.tile(t,(n, 1)).T
    
    sinc_vector = np.sinc( np.pi*(t-n_vec*dt)/dt )
    
#     return np.sum(xn * sinc_vector, axis=1)
    return xn * sinc_vector
```

```python
pi = np.pi
cos = np.cos

xn = lambda n: cos( (pi*n)/4 ) + cos( (3*pi*n)/4 )
```

# Down sampling by a factor of 4

```python
plt.figure(figsize=(16,7))

N = 48    # number of sampling points per 2pi cycle, free to choose
n = np.arange(N+1)


T = 8    # period of the minimal frequency (pi/4) in seconds this minimal amount required to plot one complete 2pi cycle
dt = T/N    # sampling rate


# original sampling points plot
plt.scatter(x=n*dt, y=xn(n*dt), label="Original sampling")


# down sampling plot
n4 = n[::4]
plt.scatter(x=n4*dt, y=xn(n4*dt), facecolors="none", edgecolor="r", s=200, label="Down sampling by 4")


# Linear interpolation
plt.plot(n4*dt, xn(n4*dt), c="r", label="Linear interpolation")


# sinc interpolation
# f = [sinc_interp(t, xn(n4*dt), dt*4) for t in n*dt]
# plt.plot(n*dt, f, c="g", label="Sinc interpolation")


plt.grid()
plt.legend()
plt.show()
```

```python
def sinc_interp(t, s, u):
    
    """
    Parameters
    ----------
    t: array-like
        Time points with fixed interval for sampled points
        
    s: array-like
        Values of sampled points, corresponding to t
        
    u: float
        Time of unsampled point at which we want to interpolate
    """
    dt = t[1] - t[0]
    
    return np.sum(s * np.sinc( (np.pi*(u - t))/dt ))
```

```python
tu = np.linspace(0.1,0.9,8)
plt.scatter(tu ,[sinc_interp(t, xt(t), ti) for ti in tu])
```
